Matplotlib reference

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
x = [1,2,3]
y = [2,4,6]

plt.plot(x, y, label="2x", color="#bbbbbb", linewidth=3, marker=".", markersize=20, linestyle="--")

plt.title("Example title", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([1,2,3])
plt.yticks([4,5,6])

plt.legend()

plt.show() #correct way to show the graph, prevents showing the memory location
```

```
x = [1,2,3]
y = [2,4,6]

plt.plot(x, y, "r.--", label="2x") #short hand notation: color, marker, line

plt.title("Example title", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([1,2,3])
plt.yticks([4,5,6])

plt.legend()

plt.show()
```

```
x = [1,2,3]
y = [2,4,6]

plt.figure(figsize=(10,5), dpi=300) #3000px x 1500px

plt.plot(x, y, "r.--", label="2x")

z = np.arange(1, 3.5, 0.5)
plt.plot(z, z**2, "g^-", label="X^2")

plt.title("Example title", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("x axis")
plt.ylabel("y axis") 

plt.legend()

#plt.savefig("test.png")

plt.show()
```

```
condition = ["Group A", "Group B", "Group C", "Group D"]
y = [10, 5, 20, 15]

bar_plot = plt.bar(condition, y)

bar_plot[0].set_hatch("/")
bar_plot[1].set_hatch("o")
bar_plot[2].set_hatch("/")
bar_plot[3].set_hatch("o")


plt.show()

```
